### Importing libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

### Load the Iris dataset

```python
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

### 1. Bar Chart: Distribution of species

```python
iris_df['species'].value_counts().plot(kind='bar', color=['skyblue', 'orange', 'green'])
plt.title('Distribution of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

### 2. Scatter Plot: Sepal Length vs Sepal Width

```python
sns.scatterplot(data=iris_df, x=iris.feature_names[0], y=iris.feature_names[1], hue='species')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

### 3. Heatmap: Correlation Matrix

```python
sns.heatmap(iris_df.iloc[:, :-1].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### 4. Box Plot: Petal Length by Species

```python
sns.boxplot(data=iris_df, x='species', y=iris.feature_names[2], palette='Set2')
plt.title('Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()
```
